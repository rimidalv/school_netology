```python
import requests
from multiprocessing.dummy import Pool
from lxml.html import fromstring

def get_page(url):
    print('url: '+url+  '\n')
    headers = {
    'User-Agent': 'Chrome',
    'From': '111@gmail.com'  # This is another valid field
    }
    req = requests.get(url, headers=headers)
    print('html: '+str(req)+  '\n')
    html = req.text
    dom = fromstring(html)
    dom.make_links_absolute(url)
    return dom

# def get_article_urls(url_date):
#     pattern = '.b-article__title a'
#     dom = get_page(url_date)
#     return [t.attrib['href'] for t in dom.cssselect(pattern)]

def get_article_text(url):
    patternText = '.text'
    patternEstimate = '.rates'
    try:
        dom = get_page(url)
    except Exception as err:
        print("OS error: {0}".format(err))
        return None, None
    
    rating = []
    try:
        d= dom.cssselect(patternText)
        text = [item.text_content() for item in d]

        dE = dom.cssselect(patternEstimate) 
        rating = [ item.get('data-r') for item in dE ]        
    except Exception as err:
        print("OS error: {0}".format(err))
        print('except css\n')
        text = [""]
        rating = [""]
        
    return rating, text[:-1]

flatten = lambda l: [item for sublist in l for item in sublist]
```

```python
url_date_template = 'https://www.anekdot.ru/release/anekdot/day/{}-{:02d}-{:02d}/' #2017-09-23#'https://www.vedomosti.ru/archive/{}/{}/{}'
url_dates = []
for y in range(2001, 2016):
    for m in range(1,13):
        for d in range(1,28):
            url_dates.append(url_date_template.format(y,m,d))
# url_dates
```

```python
df = pd.DataFrame()
```

```python
pool = Pool(1)
```

```python
%%time
url_articles = pool.map(get_article_text, url_dates)
```

```python
df["rating"] = [rat.split(';')[0] for t in url_articles for rat in t[0]]
df["text"] = [rat for t in url_articles for rat in t[1]]
```
