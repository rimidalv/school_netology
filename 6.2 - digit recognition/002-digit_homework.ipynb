### Начало взято из урока для наглядности. Ниже само решение

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Загружаем исходные данные

```python
train = np.loadtxt('./data/digit/train.csv', delimiter=',', skiprows=1)
test = np.loadtxt('./data/digit/test.csv', delimiter=',', skiprows=1)
```

```python
# сохраняем разметку в отдельную переменную
train_label = train[:, 0]
# приводим размерность к удобному для обаботки виду
train_img = np.resize(train[:, 1:], (train.shape[0], 28, 28))
test_img = np.resize(test, (test.shape[0], 28, 28))
```

```python
train_img.shape
```

## Визуализируем исходные данные

```python
def show_digits( arr ):
    fig = plt.figure(figsize=(20, 10))
    for i, img in enumerate(arr[0:5], 1):
        subplot = fig.add_subplot(1, 7, i)
        plt.imshow(img, cmap='gray');
        subplot.set_title('%s' % train_label[i - 1]);
```

```python
show_digits(train_img)
```

## Вычисляем X и Y составляющие градиента с помощью оператора Собеля

```python
def get_gradients(img):
    train_sobel_x = np.zeros_like(img)
    train_sobel_y = np.zeros_like(img)
    for i in range(len(img)):
        train_sobel_x[i] = cv2.Sobel(img[i], cv2.CV_64F, dx=1, dy=0, ksize=3)
        train_sobel_y[i] = cv2.Sobel(img[i], cv2.CV_64F, dx=0, dy=1, ksize=3)
    return train_sobel_x, train_sobel_y
```

```python
train_sobel_x, train_sobel_y = get_gradients(train_img)
test_sobel_x, test_sobel_y = get_gradients(train_img)
```

## Вычисляем угол и длину вектора градиента

```python
train_g, train_theta = cv2.cartToPolar(train_sobel_x, train_sobel_y)
```

```python
test_g, test_theta = cv2.cartToPolar(test_sobel_x, test_sobel_y)
```

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(train_g[:5], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % train_label[i - 1]);
    subplot = fig.add_subplot(3, 7, i)
    plt.hist(train_theta[i - 1].flatten(),
             bins=16, weights=train_g[i - 1].flatten())
```

## Вычисляем гистограммы градиентов

```python
# Гистограммы вычисляются с учетом длины вектора градиента
train_hist = np.zeros((len(train_img), 16))
for i in range(len(train_img)):
    hist, borders = np.histogram(train_theta[i],
                                 bins=16,
                                 range=(0., 2. * np.pi),
                                 weights=train_g[i])
    train_hist[i] = hist
```

```python
test_hist = np.zeros((len(test_img), 16))
for i in range(len(test_img)):
    hist, borders = np.histogram(test_theta[i],
                                 bins=16,
                                 range=(0., 2. * np.pi),
                                 weights=test_g[i])
    test_hist[i] = hist
```

## Нормируем вектор гистограммы

```python
# По умолчанию используется L2 норма
train_hist = train_hist / np.linalg.norm(train_hist, axis=1)[:, None]
```

```python
test_hist = test_hist / np.linalg.norm(test_hist, axis=1)[:, None]
```

## Разбиваем выборку на обучение и валидацию

```python
from sklearn.model_selection import train_test_split
y_train, y_val, x_train, x_val = train_test_split(
    train_label, train_hist, test_size=0.2,
    random_state=42)
```

## Вычисляем центроиды каждого из 10 классов

```python
centroids = np.zeros((10, train_hist.shape[1]), dtype=np.float32)
```

```python
for i in range(10):
    centroids[i] = np.mean(x_train[y_train == i], axis=0)
```

## Находим ближайший центройд для валидации

```python
pred_val = np.zeros_like(y_val)
for i in range(len(y_val)):
    distances = np.linalg.norm(centroids - x_val[i], axis=1)
    pred_val[i] = np.argmin(distances)
```

## Оцениваем качество решение на валидационной выборке

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %s' % accuracy_score(y_val, pred_val))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, pred_val))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_val, pred_val))
```

## Предсказания на тестовыйх данных

```python
pred_test = np.zeros(len(test_img), np.uint8)
for i in range(len(test_img)):
    distances = np.linalg.norm(centroids - test_hist[i], axis=1)
    pred_test[i] = np.argmin(distances)
```

## Визуализируем предсказания

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(test_img[0:5], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % pred_test[i - 1]);
```

## Готовим файл для отправки

```python
with open('submit.txt', 'w') as dst:
    dst.write('ImageId,Label\n')
    for i, p in enumerate(pred_test, 1):
        dst.write('%s,%s\n' % (i, p))
```

### Способ №1

Разделим изображения на 4 части и получим гистограммы от каждой из этих частей. Сконкантенируем их в одном векторе и обучим классификатор на этом векторе. 

```python
def get_images_quater(imgs):
    train_img_quarter = []
    for i, img in enumerate(imgs):
        train_img_quarter.append(np.append(img[ :14, :], img[ 14:, :], axis=1))
    return train_img_quarter
```

```python
train_img_quarter = get_images_quater(train_img)
test_img_quarter = get_images_quater(test_img)
```

```python
show_digits(train_img_quarter)
```

```python
train_sobel_x_quater, train_sobel_y_quater = get_gradients(train_img_quarter)
test_sobel_x_quater, test_sobel_y_quater = get_gradients(test_img_quarter)
```

```python
train_g, train_theta = cv2.cartToPolar(train_sobel_x_quater, train_sobel_y_quater)
test_g, test_theta = cv2.cartToPolar(test_sobel_x_quater, test_sobel_y_quater)
```

```python
train_g.shape
```

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(train_g[:5], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % train_label[i - 1]);
```

## Вычисляем гистограммы градиентов

```python
# Гистограммы вычисляются с учетом длины вектора градиента
def calc_hist(train_img, train_theta, train_g):
    train_hist = np.zeros((len(train_img), 64))
    for i in range(len(train_img)):
#         print(train_theta[i, :14 ].shape)
        hist1, borders = np.histogram(train_theta[i, :, :14 ],
                                     bins=16,
                                     range=(0., 2. * np.pi),
                                     weights=train_g[i, :, :14])
        hist2, borders = np.histogram(train_theta[i, :, 14:28 ],
                             bins=16,
                             range=(0., 2. * np.pi),
                             weights=train_g[i, :, 14:28 ])
        hist3, borders = np.histogram(train_theta[i, :, 28:42 ],
                                     bins=16,
                                     range=(0., 2. * np.pi),
                                     weights=train_g[i, :, 28:42])
        hist4, borders = np.histogram(train_theta[i, :, 42: ],
                             bins=16,
                             range=(0., 2. * np.pi),
                             weights=train_g[i, :, 42: ])
        train_hist[i] = np.append(np.append(np.append(hist1, hist2), hist3), hist4)
    return train_hist
```

```python
train_hist = calc_hist(train_img, train_theta, train_g)
test_hist = calc_hist(test_img, test_theta, test_g)
```

## Нормируем вектор гистограммы

```python
def get_normed(hist):
# По умолчанию используется L2 норма
    return hist / np.linalg.norm(hist, axis=1)[:, None]
```

```python
train_hist = get_normed(train_hist)
test_hist = get_normed(test_hist)
```

## Разбиваем выборку на обучение и валидацию

```python
from sklearn.model_selection import train_test_split
y_train, y_val, x_train, x_val = train_test_split(
    train_label, train_hist, test_size=0.2,
    random_state=42)
```

## Вычисляем центроиды каждого из 10 классов

```python
centroids = np.zeros((10, train_hist.shape[1]), dtype=np.float32)
```

```python
for i in range(10):
    centroids[i] = np.mean(x_train[y_train == i], axis=0)
```

## Находим ближайший центройд для валидации

```python
pred_val = np.zeros_like(y_val)
for i in range(len(y_val)):
    distances = np.linalg.norm(centroids - x_val[i], axis=1)
    pred_val[i] = np.argmin(distances)
```

## Оцениваем качество решение на валидационной выборке

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %s' % accuracy_score(y_val, pred_val))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, pred_val))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_val, pred_val))
```

## Предсказания на тестовыйх данных

```python
pred_test = np.zeros(len(test_img_quarter), np.uint8)
for i in range(len(test_img_quarter)):
    distances = np.linalg.norm(centroids - test_hist[i], axis=1)
    pred_test[i] = np.argmin(distances)
```
